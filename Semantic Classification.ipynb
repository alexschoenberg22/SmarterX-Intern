```python
#import pandas and read Excel file in
import pandas as pd

prods_xls = '1-Semantic_Classification.xlsx'
products = pd.read_excel(prods_xls)
```

```python
#remove solution columns from dataset
columns_drop = ['Unnamed: 0', 'keywords', 'Confidence', 'For_Kids?']
products = products.drop(columns_drop, axis=1)

products.head()
```

```python
#clean it up
products.columns = ['UPC', 'Name', 'Brand', 'Marketing', 'Image', 'Features']

products
```

Note: There is no Marketing Copy data (and thus no Features) for rows 1197-2195 (end).

```python
products.info()
```

```python
#dropping rows with NaN values (assuming some data was lost in the initial solution of this problem, which I dropped out column-wise to try to restart the problem)
products_complete = products.dropna()

products_complete.info()
```

```python
import numpy as np
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
from collections import Counter
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfTransformer


counter = Counter(products_complete['Name'].tolist())
top_10_categories = {i[0]: idx for idx, i in enumerate(counter.most_common(10))}
products = products_complete[products_complete['Name'].map(lambda x: x in top_10_categories)]

description_list = products_complete['Features'].tolist()
categories_list = [top_10_categories[i] for i in products_complete['Name'].tolist()]
categories_list = np.array(categories_list)

count_vect = CountVectorizer()
x_train_counts = count_vect.fit_transform(description_list)


tfidf_transformer = TfidfTransformer()
x_train_tfidf = tfidf_transformer.fit_transform(x_train_counts)

train_x, test_x, train_y, test_y = train_test_split(x_train_tfidf, categories_list, test_size=0.3)

clf = MultinomialNB().fit(train_x, train_y)
y_score = clf.predict(test_x)

n_right = 0
for i in range(len(y_score)):
    if y_score[i] == test_y[i]:
        n_right += 1

print("Accuracy: %.2f%%" % ((n_right/float(len(test_y)) * 100)))
```
